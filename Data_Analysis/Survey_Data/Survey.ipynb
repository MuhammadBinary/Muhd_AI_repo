```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Survey Data.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={
    'How old are you?': 'Age',
    'What industry do you work in?': 'Industry',
    'If your job title needs additional context, please clarify here:': 'Job_title(add_content)',
    "What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)": "Annual_Salary",
    "How much additional monetary compensation do you get, if any (for example, bonuses or overtime in an average year)? Please only include monetary compensation here, not the value of benefits.": 'Monetary_compensation',
    "Please indicate the currency": 'Currency',
    'If "Other," please indicate the currency here: ': 'Currency(if others)',
    "If your income needs additional context, please provide it here:": "income needs",
    "What country do you work in?": "job_country",
    "If you're in the U.S., what state do you work in?": "US(state_job)",
    "What city do you work in?":"city",
    "How many years of professional work experience do you have overall?": "Prof.Work_Experience(overall)",
    "How many years of professional work experience do you have in your field?": "Prof.Work_Experience(field only)",
    "What is your highest level of education completed?": "Highest_Level(Education)",
    "What is your gender?": "Gender",
    "What is your race? (Choose all that apply.)": "race"
})
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Job title'].value_counts()
```

```python
df['Monetary_compensation'] = df['Monetary_compensation'].fillna(df['Monetary_compensation'].mode()[0])
```

```python
df['Job_title(add_content)'] = df['Job_title(add_content)'].fillna('others')
df['income needs'] = df['income needs'].fillna('others')
df['US(state_job)'] = df['US(state_job)'].fillna('California')
df['city'] = df['city'].fillna('Boston')
df['Highest_Level(Education)'] = df['Highest_Level(Education)'].fillna('College degree')
df['race'] = df['race'].fillna('White')
df['Currency(if others)'] = df['Currency(if others)'].fillna('others')
df['Job title'] = df['Job title'].fillna('Software Engineer')
df['Age'] = df['Age'].fillna('Woman')

df.isnull().sum()
```

```python
df[['date','time']] = df['Timestamp'].str.split(' ', expand=True)
del df['Timestamp']
```
