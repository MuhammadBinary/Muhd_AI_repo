```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Tennis_dataset.csv')
# before = pd.read_csv('Tennis_dataset.csv')
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder

order_map = [['Overcast', 'Rainy', 'Sunny']]
order_map1 = [['Mild', 'Cool', 'Hot']]
order_map2 = [['Normal', 'High']]


encoder = OrdinalEncoder(categories=order_map)
encoded_df = encoder.fit_transform(df[['Outlook']])
df['Outlook'] = encoded_df

encoder1 = OrdinalEncoder(categories=order_map1)
encoded_df1 = encoder1.fit_transform(df[['Temperature']])
df['Temperature'] = encoded_df1

encoder2 = OrdinalEncoder(categories=order_map2)
encoded_df2 = encoder2.fit_transform(df[['Humidity']])
df['Humidity'] = encoded_df2

encoder3 = LabelEncoder()
encoded_df3 = encoder3.fit_transform(df['Windy'])
df['Windy'] = encoded_df3

df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

X = df.drop('PlayTennis', axis=1)
y = df['PlayTennis']

dtree = DecisionTreeClassifier()
dtree.fit(X, y)
pred = dtree.predict([[2.0, 2.0, 1.0, 1]])
pred
```
